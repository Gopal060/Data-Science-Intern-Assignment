### **Import the Required Libraries**

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

### **Load the datasets**

```python
customers_df = pd.read_csv('Customers.csv')
products_df = pd.read_csv('Products.csv')
transactions_df = pd.read_csv('Transactions.csv')
```

```python
# Cheking customers data and info about data
customers_df.head()
```

```python
# first 5 records from products
products_df.head()
```

```python
# first 5 records from transactions
transactions_df.head()
```

```python
# merging datafrmaes based on the column between transactions and products and then products and customers. 

merge_df = transactions_df.merge(products_df, on='ProductID').merge(customers_df, on='CustomerID')
merge_df
```

```python
# First 5 rows form the merge dataframe

merge_df.head()
```

```python
# first checking the missing values 

missing_values = merge_df.isnull().sum()
missing_values
```

```python
# There is no missing values present in the dataset
```

```python
## basic summery of the dataset by using describe


summery_info = merge_df.describe()
summery_info
```

```python
merge_df.info()
```

```python
merge_df.head()
```

```python
# Convert TransactionDate to datetime for time-based analysis (currently it is object data type)

merge_df['TransactionDate'] = pd.to_datetime(merge_df['TransactionDate'])
```

```python
# check data type converted or not
merge_df.info()
```

```python
# Extract Month-Year from TransactionDate

merge_df['MonthYear'] = merge_df['TransactionDate'].dt.to_period('M')
```

```python
# Final dataframe containing all necessary information for exploratory data analysis.
merge_df.head()
```

### **Exploratory Data Analysis**

```python
# As we know we have category column so first check with category wise analysis

category_sales = merge_df.groupby('Category')['TotalValue'].sum().sort_values(ascending=False)
```

```python
category_sales
```

```python
# Top Selling Products

top_selling_products = merge_df.groupby('ProductName')['TotalValue'].sum().sort_values(ascending=False).head()
top_selling_products
```

```python
# Region wise sales 

region_wise_sales = merge_df.groupby('Region')['TotalValue'].sum().sort_values(ascending=False)
region_wise_sales
```

```python
# Monthly sales trends about products

monthly_sales_product = merge_df.groupby('MonthYear')['TotalValue'].sum()
monthly_sales_product
```

```python
# category wise sales visualization

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.barplot(x=category_sales.index, y=category_sales.values, palette='viridis')
plt.xlabel("Total Sales by Category")
plt.ylabel("Category")
plt.xticks()
plt.show()
```

```python
# Monthly sales visualization

plt.figure(figsize=(10, 5))
monthly_sales_product.plot(marker='o', color='teal')
plt.title("Monthly Sales Trend")
plt.xlabel("Month-Year")
plt.ylabel("Total Sales Value")
plt.grid()
plt.show()
```
